```python
def get_cmap_rchi2(vmin, vmax):
    orig_cmap = matplotlib.cm.RdBu_r
    start, stop = get_points_for_colormap(vmin, vmax, central_val=1.)
    midpoint = (1 - vmin) / (vmax - vmin)
    return shiftedColorMap(orig_cmap, start=0., midpoint=midpoint, stop=stop)


def add_style(ax):
    ax.set_xlabel('Galactic Longitude')
    ax.set_ylabel('Galactic Latitude')


if not os.path.exists('decomposition_grs'):
    !mkdir decomposition_grs
```

**6. Based on the $\chi^2$ map, did the decomposition perform well? How many components did Gausspy find on average?**

Based on the $\chi^{2}$ map, it can be observed plenty of pixels whose number are between 1 and 1.75 i.e. there are plenty of pixels with colors from white to red light, while there are few pixels whose numbers were high i.e. 2.50 (intense red color). Thus, the decomposition performs well.

As we can observe from the plot about the Number of fitted components, we can infer that Gausspy found on average 3 components.

However, it is possible to make a numerical analysis about how many Gaussian components was found on average.

```python
# read the data file

data_dir = './decomposition_grs/gpy_decomposed/'
data_file = 'cloud_sub_cube_g+_fit_fin_sf-p2_finalized.dat'
data = ascii.read(os.path.join(data_dir + data_file))
```

```python
data.keys()
```

```python
max_x = np.max(data["x_pos"])
max_y = np.max(data["y_pos"])
```

```python
print("The max position for x is: ", max_x)
print("The max position for y is: ", max_y)
```

```python
g_components=[]

for x_pix in range(max_x+1):
    #print(x_pix)
    for y_pix in range(max_y+1):
        
        data_2 = data[data['x_pos']==x_pix]
        data_3 = data_2[data_2['y_pos']==y_pix]
        
        gauss_component = len(data_3)
        
        g_components.append(gauss_component)
```

```python
g_components = np.array(g_components)

average =  np.mean(g_components)
print("Finally, Gausspy find on average", average, "components")

average_2 = int(average)
print("So, if we approximate to the integer number, around:", average_2 )
```

**7. Make histograms of the decomposition results for $v_{LSR}$ , FWHM (full width at half max), and amplitude. What can you conclude about the structure of the interstellar gas based on the histograms?**

**Resource: Analyse_decomposition_results.ipynb**


```python
# plot velocity distributio of the components
# this indicates the emission at different velocities -> motion of the gas + different clouds in the line of sight

fig = plt.figure(figsize=(8.,5.))
plt.axes(frameon=False)
plt.xticks([])
plt.yticks([])

ax = fig.add_subplot(1,1,1)
d = data['VLSR']
#bin_size = range(80,200, 2) 
plt.hist(d, bins=200, histtype='step', color='C0', fill=True, linewidth=2)
plt.xticks(family='serif',size=18)
plt.yticks(family='serif',size=18)
ax.set_xlabel(r'$v_{LSR}$ [km s$^{-1}$]', size=20,family='serif')
ax.set_ylabel('N', size=20,family='serif')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# FWH plots

fig = plt.figure(figsize=(6.,5.))
plt.axes(frameon=False)
plt.xticks([])
plt.yticks([])

ax = fig.add_subplot(1,1,1)

#convert sigma to FWHM:
d = abs(data['vel_disp'])*2.355

plt.hist(d, bins=100, histtype='step', color='C1', fill=True, linewidth=2)
plt.xticks(family='serif',size=18)
plt.yticks(family='serif',size=18)
ax.set_xlabel(r'$v_{FWHM}$ [km s$^{-1}$]', size=20,family='serif')
ax.set_ylabel('N', size=20,family='serif')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# distribution of amplitude

fig = plt.figure(figsize=(6.,5.))
plt.axes(frameon=False)
plt.xticks([])
plt.yticks([])

ax = fig.add_subplot(1,1,1)
d = abs(data['amp'])
plt.hist(d, bins=100, histtype='step', color='C2', fill=True, linewidth=2)

plt.xticks(family='serif',size=18)
plt.yticks(family='serif',size=18)
ax.set_xlabel(r'$Amplitude$', size=20,family='serif')
ax.set_ylabel('N', size=20,family='serif')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Based on the $v_{LSR}$ histogram, we can conclude that the interestellar gas are not moving at relative same velocity in the overall structure. 

Based on the $v_{FWHM}$ histogram, we can conclude that the difference between the velocities are between 0 $km s^{-1}$ and 40 $km s^{-1}$. It goes according with the difference of velocities in $v_{LSR}$ because the most difference is between 110 and 150 $v_{LSR}$ $[km s^{-1}]$ 

Based on the Amplitude histogram, we can conclude that the most of the cloud has a relation between 10 and 15. Since it is related with the radial velocity we can conclude that the cloud is rotating.

The difference in velocities for the overall gas cloud means that it goes in a relative rotational motion between the SMC. 

Also, the chemical composition is not the same and it would explains why the gas move at different velocities.

**6. What do you think is the reason for the differences between your fit and the results of Güver and Özel 2009?**

This discrepancy is due to the model just considering the column density made up of neutral Hydrogen N(HI). So, it can be inferred that not all column density composition is HI. Actually, it is $H_{2}$ 

The difference is because of the concentration of N(HI) that would appear to stabilize, as we can observe in the fitting plot where there is a stable concentration at 1.0 $N(HI) [atoms/cm^{2}]$ that extends for more values of $A_{v} [mag]$. The reason for this is that dust grains enhance the production of $H_{2}$ (more than it would be expected by random collisions of Hydrogen atoms) and other molecules known as **tracers of $H_{2}$** because its abundances are related with $H_{2}$ abundance.
