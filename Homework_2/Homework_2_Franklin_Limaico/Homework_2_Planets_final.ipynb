**2. Make sure that you only have relevant data for planets. For this, think about the definition of a planet and exclude all data that does not satisfy the definition. Briefly describe your decision about excluding irrelevant data.**


The relevant data for planets are based on the mass. Since Brown Dwarfs are considered to have masses between two limits ($13 M_J < M_{bd}< 75 M_J$), so the relevant data for planets are those ones have masses less than thirteen Jupyter Mases (i.e. $<13M_J$)


```python
#print(data["mass"])

indices = np.where(data["mass"]>13)

# Deleting elements in the selected indices
isolated_mass = np.delete(data["mass"], indices)

# Finding indices where the value is NaN
indices_nans = np.isnan(isolated_mass)

# Deleting NaN elements
planets = isolated_mass[~indices_nans]

# Converting the array in a DataFrame
planets = pd.DataFrame(planets)

print("Data frame with the data excluded according to the criteria of <13 Jupyter masses:", planets)
```

```python
#data.keys()
```

```python
#Define the volume of the exoplanets based on the radius of the exoplanets
radius_array = data['radius']

volume_array = (4/3)*(np.pi)*(radius_array**3)

#Now, define the mass of the exoplanets:
mass_array = data['mass']

#Now, obtain the density of the exoplanets as array:

density_array = mass_array/volume_array
dens = data['mass']/((4/3)*(np.pi)*(data['radius']**3))
print(density_array)
```
